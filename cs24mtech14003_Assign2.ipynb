```python
# Programing question Q4. a)

# Here, i am importing numpy for numerical operation.
import numpy as np
# As the accuracy_score name suggest, i am using this to compute models accuracy.
from sklearn.metrics import accuracy_score
# The motivation behind, to importing SVC from SVM which is present inside sklearn is to train and creat the SVM model.
from sklearn.svm import SVC
# Here,i am impoting pandas for data manipulation and for rading csv files from online source and also for analysis.
import pandas as pd
# from sklearn import *

# Here, i am defining the URLs where the actual taingning and testing data is present or loacted.
trn = 'http://www.amlbook.com/data/zip/features.train'
tst = 'http://www.amlbook.com/data/zip/features.test'

# Here, i am Loadig the data sets using the pandas DataFrames
# the training_data variable, represents the entire taining data set. 
training_data = pd.read_csv(trn, header=None, sep = '\s+')
# the testing_data variable, represents the entire testing data set. 
testing_data = pd.read_csv(tst, header=None, sep = '\s+')

# The following two line of code is used to Filter the dataset for digits '1' and '5' in coloum 1.
# here i am selecting the coloum one's value form training data set.
t = training_data[0].values
# then on slected coloum, i am only taking those rows whose value is either 1 or 5.
tr = np.isin(t, [1, 5])
# and afte this assigning the filtered rows into a tain_data variable.
trn_data = training_data[tr]

# here i am selecting the coloum one's value form testing data set.
t = testing_data[0].values
# then on slected coloum, i am only taking those rows whose value is either 1 or 5.
ts = np.isin(t, [1,5])
# and afte this, assigning the filtered rows into a test_data variable.
tst_data = testing_data[ts]

# extracting features and labels
# from filtered training data set extracting featrutres and labels.
X_trn = trn_data[[1, 2]].values
y_trn = trn_data[0].values
# from filtered testing data set extracting featrutres and labels.
X_tst = tst_data[[1, 2]].values
y_tst = tst_data[0].values

# here we are working on the binary classification problem
"""so, in the below lines of code, for both test and training dataset
i am assigning 1 as +1 and 5 as -1"""
y_trn_nn = []
for i in y_trn:
    if i == 1:
        y_trn_nn.append(1)
    else:
        y_trn_nn.append(-1)
y_trn = y_trn_nn

y_tst_nn = []
for i in y_tst:
    if i == 5:
        y_tst_nn.append(-1)
    else:
        y_tst_nn.append(1)
y_tst = y_tst_nn

# uisng linear kernal, i am training the SVM model.
# C is the regularization parameter in Soft margin SVM.
model = SVC(kernel='linear', C=1.0)  
model.fit(X_trn, y_trn)

# here i am using training SVM model to make pediction on labels for the test data.
y_pred = model.predict(X_tst)

# Q4. a)
# here i am trying to calculate the number of suport vector in soft margine svm
sup_vec = model.support_vectors_
num_support_vectors = len(sup_vec)

# now we are comparing the predicted labels with the true labels, to compute the accuracy of a model.
accurcy = accuracy_score(y_tst, y_pred)

print("Assignment 2\n")
print("Programming Question Q4. a)")
print(f'  Accuracy for the test set : {accurcy:.4f}')
print(f'  Number of support vectors by using the soft margin svm : {num_support_vectors}')



# Q4. b)
# Here i am defining empty dictionary, which will store the results 
rslt = {}

# the lsit define below is the number of points that is used for training.
trn_sz = [50, 100, 200, 800]

# this loop will for each specified training size.
for i in trn_sz:
    # here it will initialize svm models linear kernal and regularization parameter.
    model = SVC(kernel='linear', C=1.0)

    # here it will extract training sample from the data set.
    X_train = X_trn[:i]
    y_train = y_trn[:i]
    
    # it will fits model to the training data.
    model.fit(X_train, y_train)
    
    # it will going to predect model with the test data for each test case.
    y_pred = model.predict(X_tst)
    
    # it will comparing the predected values with the true values to find the value of accuracy.
    accur = accuracy_score(y_tst, y_pred)
    
    # it will extracting the number of suppoet vector which is used by the model.
    numb_vect = len(model.support_vectors_)
    
    # distionary with name result, which will going to store accuracy and number of vectors for each test case.
    rslt[i] = {
        'accuracy': accur,
        'num_support_vectors': numb_vect
    }

print("\nPrograming Question Q4. b)")

for i in trn_sz:
    print(f"  Training the model with {i} points:")
    print(f"    Accuracy with this test set is: {rslt[i]['accuracy']:.4f}")
    print(f"    Number of support vectors in this case is: {rslt[i]['num_support_vectors']}")
    print()



# Q4. c)
print("Programming Question Q4. c)")

# value of c hyperparameter value and degree is given in the question, we have to solve accordingely.
c_val = [0.0001, 0.001, 0.01, 1]
deg = [2,5]

# here we are defining the which will help up to evaluate models performance.
# by calculating testing error, test error and number of support vectors.
def modl_evaluate(C, degree):
    modl = SVC(kernel = 'poly', degree = degree, C=C)

    modl.fit(X_trn, y_trn)

    y_p_trn = modl.predict(X_trn)
    y_p_tst = modl.predict(X_tst)

    # it will finds training and testing accuracy by uisng its own dataset.
    accuracy_of_trainig = accuracy_score(y_trn, y_p_trn)
    accuracy_of_testing = accuracy_score(y_tst, y_p_tst)

    # lets find training and teting error by subtracting 1 and accuracy, which we have find in above steps.
    tst_error = 1 - accuracy_of_testing
    trn_error = 1 - accuracy_of_trainig

    # now finding number of suport vector in this case, which will helps the model to evaluate the accuarcy and error.
    numb_of_sup_vec = len(modl.support_vectors_)

    # this function will returns the traning error, testing error and number of support vector.
    return trn_error, tst_error, numb_of_sup_vec

result = {}
# now the following loop will run for all hyper parameters and degree of a polynomial.
for i in c_val:
    for j in deg:
        trn_error, tst_error, numb_of_sup_vec = modl_evaluate(i, j)
        # for each value of c and degree, model is evaluating training error, testing error and support vector, and store this in dictionary.
        result[(i, j)]={'training_error':trn_error,
                        'testing_error':tst_error,
                        'support_vector':numb_of_sup_vec
                        }
        

print("this is that what i get by using hyperparameter c and polynomial degree.\n")

# the following degree is used to print values of error with respect to each hyperparameter and degree of a polynomail.
def printing(c, degree, matrix):
    print(
        f"On polynomial degree: {degree}, c: {c}\n"+
        f"    Trianing error is: {matrix['training_error']}\n"
        f"    Testing error is: {matrix['testing_error']}\n"
        f"    Number of support vector for this case is: {matrix['support_vector']}\n"
    )

# this itterate for each element of dictionay.
for (i, j), matrics in result.items():
    printing(i, j, matrics)

# now evaluating statement, as per given in the question.
print("Here, i am evaluating the given statements")

# here i am initialising the nested dictionary, to saperately store the training error, testing error and support vector for the given values of degree.
train_error = {}
for i in deg:
    train_error[i] = {}

test_error = {}
for i in deg:
    test_error[i] = {}

vectors = {}
for i in deg:
    vectors[i] = {}

# now inserting the values of error and support vector in the empty nested dictionary.
for i in c_val:
    for j in deg:
        vectors[j][i] = result[(i, j)]['support_vector']
        train_error[j][i] = result[(i, j)]['training_error']
        test_error[j][i] = result[(i, j)]['testing_error']
        

statement_i = train_error[5][0.0001] > train_error[2][0.0001]
print(f"    Statement (i). {statement_i}")

statement_ii = vectors[2][0.001] > vectors[5][0.001]
print(f"    Statement (ii). {statement_ii}")

statement_iii  = vectors[5][0.01] > vectors[2][0.01]
print(f"    Statement (iii). {statement_iii}")

statement_iv = vectors[2][1] > vectors[5][1]
print(f"    Statement (iv). {statement_iv}")


# Q4. d)
print("\nProgramming Question Q4. d)")
# here we have to work on Radial Basis Function Kernal
C_val = [0.01, 1.0, 100.0, 10000.0, 1000000.0]

# lets create a empty dectionary, which will stores the answers.
rslt = {}

# here i am defining the function which helps me to evaluate the performance of a model.
def eval_model(c):
    # here i am specifing that i am using an an RBF kernal.
    modl = SVC(kernel = 'rbf', C=c, gamma = 'scale')
    # in this line i, am trying to fit train dataset into the model
    modl.fit(X_trn, y_trn)

    # now i am making prediction on train and test dataset.
    y_trn_prediction = modl.predict(X_trn)
    y_tst_prediction = modl.predict(X_tst)

    # to calculate training error and testing error, for that i have to calculate accuray first.
    tst_accur = accuracy_score(y_tst, y_tst_prediction)
    trn_accur = accuracy_score(y_trn, y_trn_prediction)

    # now subtracting 1 with accuracy value to get the error.
    trn_err = 1 - trn_accur
    tst_err = 1 - tst_accur

    return tst_err, trn_err

# this loop will run till number element in C_val.
for i in C_val:
    tst_err, trn_err = eval_model(i)
    # this dictionary will store training error and testing error for each value of C_val.
    rslt[i] = {
        'trn_error': trn_err,
        'tst_error': tst_err
    }

# now printing all the values of a result for each values of C_val.
print(f"{'C':<10} {'Testing Error':<17} {'Traning Error'}")
print("----------------------------------------------")

# we have to find lowest training error and testing error.
testing_error = float('inf')
training_error = float('inf')

for i in C_val:
    print(f"{i:<10} {rslt[i]['tst_error']:<17.5f} {rslt[i]['trn_error']:.5f}")
    if testing_error > rslt[i]['tst_error']:
        testing_error = rslt[i]['tst_error']
    if training_error > rslt[i]['trn_error']:
        training_error = rslt[i]['trn_error']
# print(testing_error)
# print(training_error)
for i in C_val:
    if testing_error == rslt[i]['tst_error']:
        print(f"\nLowest testing error at C : {i} and the error value : {testing_error:.5f}")
        break

for i in C_val:
    if training_error == rslt[i]['trn_error']:
        print(f"Lowest training error at C : {i} and the error value : {training_error:.5f}")
        break


```
